```
# üöÄ SIMULA√á√ÉO COMPLETA DE ATTENTION EM TRANSFORMERS
# üìö Tutorial Educacional Interativo - Vers√£o Melhorada
# üéØ Baseado nos conceitos fundamentais de Deep Learning sobre Transformers

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.special import softmax
import pandas as pd
from matplotlib.patches import Rectangle
import warnings
warnings.filterwarnings('ignore')

# ‚ú® Configura√ß√£o visual aprimorada
plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['font.size'] = 10
plt.rcParams['axes.titlesize'] = 12
plt.rcParams['axes.labelsize'] = 10
sns.set_palette("husl")

def print_header(title, emoji="üî•", width=80):
    """Fun√ß√£o para criar cabe√ßalhos visuais elegantes"""
    print("\n" + "=" * width)
    print(f"{emoji} {title.upper()} {emoji}")
    print("=" * width)

def print_step(step_num, title, emoji="üìå"):
    """Fun√ß√£o para marcar cada etapa do processo"""
    print(f"\n{emoji} ETAPA {step_num}: {title}")
    print("‚îÄ" * (len(title) + 15))

def print_info(text, emoji="üí°"):
    """Fun√ß√£o para destacar informa√ß√µes importantes"""
    print(f"{emoji} {text}")

def print_math_explanation(formula, explanation):
    """Fun√ß√£o para explicar f√≥rmulas matem√°ticas"""
    print(f"üßÆ F√ìRMULA: {formula}")
    print(f"üìù EXPLICA√á√ÉO: {explanation}")

# =============================================================================
# üèóÔ∏è CLASSE PRINCIPAL DO SIMULADOR DE TRANSFORMERS
# =============================================================================

class TransformerSimulator:
    def __init__(self, d_model=64, seq_length=8, vocab_size=1000):
        """
        üéØ Inicializa o simulador de Transformer

        Par√¢metros:
        - d_model: Dimens√£o do modelo (tamanho dos embeddings)
        - seq_length: Comprimento da sequ√™ncia de entrada
        - vocab_size: Tamanho do vocabul√°rio
        """
        self.d_model = d_model
        self.seq_length = seq_length
        self.vocab_size = vocab_size

        print_info(f"Inicializando Transformer com dimens√£o {d_model}")
        print_info(f"Sequ√™ncia de comprimento {seq_length}")

        # üé≤ Inicializar pesos aleat√≥rios para as matrizes W_Q, W_K, W_V
        np.random.seed(42)  # Para resultados reproduz√≠veis
        self.W_q = np.random.randn(d_model, d_model) * 0.1
        self.W_k = np.random.randn(d_model, d_model) * 0.1
        self.W_v = np.random.randn(d_model, d_model) * 0.1

        print_info("Matrizes de peso inicializadas com distribui√ß√£o normal")

    def create_positional_encoding(self):
        """
        üéµ Cria encoding posicional sinusoidal

        O encoding posicional permite ao modelo entender a ordem dos tokens
        sem usar recorr√™ncia, usando fun√ß√µes seno e cosseno de diferentes frequ√™ncias.
        """
        print_step(1, "CRIANDO POSITIONAL ENCODING", "üéµ")

        pe = np.zeros((self.seq_length, self.d_model))

        print_math_explanation(
            "PE(pos,2i) = sin(pos/10000^(2i/d_model))",
            "Posi√ß√µes pares usam fun√ß√£o seno"
        )
        print_math_explanation(
            "PE(pos,2i+1) = cos(pos/10000^(2i/d_model))",
            "Posi√ß√µes √≠mpares usam fun√ß√£o cosseno"
        )

        for pos in range(self.seq_length):
            for i in range(0, self.d_model, 2):
                # Aplicar seno para dimens√µes pares
                pe[pos, i] = np.sin(pos / (10000 ** (i / self.d_model)))
                # Aplicar cosseno para dimens√µes √≠mpares
                if i + 1 < self.d_model:
                    pe[pos, i + 1] = np.cos(pos / (10000 ** (i / self.d_model)))

        print_info("Encoding posicional criado com sucesso!")
        print_info(f"Forma: {pe.shape} (posi√ß√µes √ó dimens√µes)")

        return pe

    def create_sample_embeddings(self):
        """
        üìù Cria embeddings de exemplo para demonstra√ß√£o

        Simula embeddings para a frase: "O gato subiu no telhado ontem √† noite"
        Cada token √© representado por um vetor denso de dimens√£o d_model.
        """
        print_step(2, "CRIANDO EMBEDDINGS DE TOKENS", "üìù")

        # Tokens da nossa frase de exemplo
        tokens = ["O", "gato", "subiu", "no", "telhado", "ontem", "√†", "noite"]
        print_info(f"Frase de exemplo: {' '.join(tokens)}")

        # Gerar embeddings base aleat√≥rios
        embeddings = np.random.randn(self.seq_length, self.d_model) * 0.5

        # üé® Adicionar estrutura sem√¢ntica simulada para diferentes tipos de palavras
        print_info("Adicionando caracter√≠sticas sem√¢nticas simuladas:")

        modifications = [
            ("O", "artigo - reduzindo magnitude", 0.8),
            ("gato", "substantivo - aumentando", 0.3),
            ("subiu", "verbo - aumentando", 0.5),
            ("no", "preposi√ß√£o - reduzindo", 0.7),
            ("telhado", "substantivo - aumentando", 0.4),
            ("ontem", "adv√©rbio temporal - aumentando", 0.2),
            ("√†", "preposi√ß√£o - reduzindo", 0.6),
            ("noite", "substantivo - aumentando", 0.1)
        ]

        for i, (token, desc, factor) in enumerate(modifications):
            if factor < 1:
                embeddings[i] *= factor
            else:
                embeddings[i] += factor
            print(f"  ‚Ä¢ {token}: {desc}")

        print_info(f"Embeddings criados com forma: {embeddings.shape}")
        return embeddings

    def visualize_embeddings_and_positional(self):
        """
        üìä Visualiza embeddings e encoding posicional de forma educativa
        """
        print_header("VISUALIZA√á√ÉO DE EMBEDDINGS E POSITIONAL ENCODING", "üìä")

        fig, axes = plt.subplots(2, 2, figsize=(18, 12))
        fig.suptitle('üß† Fundamentos dos Transformers: Embeddings e Posi√ß√µes',
                     fontsize=16, fontweight='bold')

        # Gerar dados
        embeddings = self.create_sample_embeddings()
        pos_encoding = self.create_positional_encoding()
        tokens = ['O', 'gato', 'subiu', 'no', 'telhado', 'ontem', '√†', 'noite']

        # 1Ô∏è‚É£ Embeddings de tokens
        im1 = axes[0,0].imshow(embeddings.T, cmap='RdBu_r', aspect='auto', interpolation='nearest')
        axes[0,0].set_title('1Ô∏è‚É£ Token Embeddings\n(Representa√ß√£o Sem√¢ntica)', fontweight='bold')
        axes[0,0].set_xlabel('Posi√ß√£o na Sequ√™ncia', fontweight='bold')
        axes[0,0].set_ylabel('Dimens√µes do Embedding', fontweight='bold')
        axes[0,0].set_xticks(range(8))
        axes[0,0].set_xticklabels(tokens, rotation=45, ha='right')
        axes[0,0].grid(True, alpha=0.3)
        plt.colorbar(im1, ax=axes[0,0], label='Valor do Embedding')

        # 2Ô∏è‚É£ Positional Encoding
        im2 = axes[0,1].imshow(pos_encoding.T, cmap='viridis', aspect='auto', interpolation='nearest')
        axes[0,1].set_title('2Ô∏è‚É£ Positional Encoding\n(Informa√ß√£o de Posi√ß√£o)', fontweight='bold')
        axes[0,1].set_xlabel('Posi√ß√£o na Sequ√™ncia', fontweight='bold')
        axes[0,1].set_ylabel('Dimens√µes do Encoding', fontweight='bold')
        axes[0,1].set_xticks(range(8))
        axes[0,1].set_xticklabels(range(8))
        axes[0,1].grid(True, alpha=0.3)
        plt.colorbar(im2, ax=axes[0,1], label='Valor do Encoding')

        # 3Ô∏è‚É£ Embeddings finais (combinados)
        final_embeddings = embeddings + pos_encoding
        im3 = axes[1,0].imshow(final_embeddings.T, cmap='RdBu_r', aspect='auto', interpolation='nearest')
        axes[1,0].set_title('3Ô∏è‚É£ Embeddings Finais\n(Sem√¢ntica + Posi√ß√£o)', fontweight='bold')
        axes[1,0].set_xlabel('Posi√ß√£o na Sequ√™ncia', fontweight='bold')
        axes[1,0].set_ylabel('Dimens√µes Finais', fontweight='bold')
        axes[1,0].set_xticks(range(8))
        axes[1,0].set_xticklabels(tokens, rotation=45, ha='right')
        axes[1,0].grid(True, alpha=0.3)
        plt.colorbar(im3, ax=axes[1,0], label='Valor Final')

        # 4Ô∏è‚É£ Padr√µes sinusoidais detalhados
        colors = plt.cm.tab10(np.linspace(0, 1, 10))
        for i in range(min(10, self.d_model)):
            axes[1,1].plot(pos_encoding[:, i], color=colors[i],
                          label=f'Dim {i}', linewidth=2, alpha=0.8)

        axes[1,1].set_title('4Ô∏è‚É£ Padr√µes Sinusoidais\n(Diferentes Frequ√™ncias)', fontweight='bold')
        axes[1,1].set_xlabel('Posi√ß√£o na Sequ√™ncia', fontweight='bold')
        axes[1,1].set_ylabel('Valor do Encoding', fontweight='bold')
        axes[1,1].legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        axes[1,1].grid(True, alpha=0.3)

        plt.tight_layout()
        plt.show()

        # üìà Estat√≠sticas educativas
        print_info("ESTAT√çSTICAS DOS EMBEDDINGS:")
        print(f"  ‚Ä¢ Embedding m√©dio: {np.mean(embeddings):.4f}")
        print(f"  ‚Ä¢ Desvio padr√£o: {np.std(embeddings):.4f}")
        print(f"  ‚Ä¢ Encoding posicional m√©dio: {np.mean(pos_encoding):.4f}")
        print(f"  ‚Ä¢ Embedding final m√©dio: {np.mean(final_embeddings):.4f}")

        return final_embeddings

    def compute_attention_step_by_step(self, X):
        """
        üîç Computa self-attention passo a passo com explica√ß√µes detalhadas
        """
        print_header("MECANISMO DE SELF-ATTENTION PASSO A PASSO", "üîç")

        tokens = ['O', 'gato', 'subiu', 'no', 'telhado', 'ontem', '√†', 'noite']

        # üéØ PASSO 1: Criar matrizes Query, Key e Value
        print_step(1, "CRIANDO MATRIZES QUERY, KEY E VALUE", "üéØ")

        print_math_explanation(
            "Q = X @ W_q,  K = X @ W_k,  V = X @ W_v",
            "Cada token gera uma query (o que procura), key (o que oferece) e value (informa√ß√£o a ser passada)"
        )

        Q = X @ self.W_q  # Queries - "O que cada token est√° procurando?"
        K = X @ self.W_k  # Keys - "O que cada token oferece?"
        V = X @ self.W_v  # Values - "Que informa√ß√£o cada token carrega?"

        print_info("DIMENS√ïES DAS MATRIZES:")
        print(f"  üì• Input (X): {X.shape} - {X.shape[0]} tokens √ó {X.shape[1]} dimens√µes")
        print(f"  üîç Queries (Q): {Q.shape} - Perguntas de cada token")
        print(f"  üîë Keys (K): {K.shape} - Respostas oferecidas por cada token")
        print(f"  üíé Values (V): {V.shape} - Informa√ß√µes a serem transmitidas")

        # Visualiza√ß√£o das matrizes Q, K, V
        fig, axes = plt.subplots(1, 3, figsize=(20, 6))
        fig.suptitle('üîç Matrizes Query, Key e Value', fontsize=16, fontweight='bold')

        # Query Matrix
        im1 = axes[0].imshow(Q.T, cmap='Reds', aspect='auto', interpolation='nearest')
        axes[0].set_title('üîç Query Matrix (Q)\n"O que cada token procura?"', fontweight='bold')
        axes[0].set_xlabel('Tokens', fontweight='bold')
        axes[0].set_ylabel('Dimens√µes Query', fontweight='bold')
        axes[0].set_xticks(range(8))
        axes[0].set_xticklabels(tokens, rotation=45, ha='right')
        axes[0].grid(True, alpha=0.3)
        plt.colorbar(im1, ax=axes[0], label='Intensidade Query')

        # Key Matrix
        im2 = axes[1].imshow(K.T, cmap='Greens', aspect='auto', interpolation='nearest')
        axes[1].set_title('üîë Key Matrix (K)\n"O que cada token oferece?"', fontweight='bold')
        axes[1].set_xlabel('Tokens', fontweight='bold')
        axes[1].set_ylabel('Dimens√µes Key', fontweight='bold')
        axes[1].set_xticks(range(8))
        axes[1].set_xticklabels(tokens, rotation=45, ha='right')
        axes[1].grid(True, alpha=0.3)
        plt.colorbar(im2, ax=axes[1], label='Intensidade Key')

        # Value Matrix
        im3 = axes[2].imshow(V.T, cmap='Blues', aspect='auto', interpolation='nearest')
        axes[2].set_title('üíé Value Matrix (V)\n"Que informa√ß√£o cada token carrega?"', fontweight='bold')
        axes[2].set_xlabel('Tokens', fontweight='bold')
        axes[2].set_ylabel('Dimens√µes Value', fontweight='bold')
        axes[2].set_xticks(range(8))
        axes[2].set_xticklabels(tokens, rotation=45, ha='right')
        axes[2].grid(True, alpha=0.3)
        plt.colorbar(im3, ax=axes[2], label='Intensidade Value')

        plt.tight_layout()
        plt.show()

        # üßÆ PASSO 2: Calcular Attention Scores
        print_step(2, "CALCULANDO ATTENTION SCORES", "üßÆ")

        print_math_explanation(
            "Attention_Scores = Q @ K^T",
            "Produto escalar entre queries e keys - mede compatibilidade entre tokens"
        )

        attention_scores = Q @ K.T
        print_info(f"Attention Scores calculados: {attention_scores.shape}")
        print_info("Cada elemento [i,j] representa qu√£o relevante o token j √© para o token i")

        # Aplicar escala
        print_math_explanation(
            "Scaled_Scores = Attention_Scores / ‚àöd_k",
            f"Divis√£o por ‚àö{self.d_model} = {np.sqrt(self.d_model):.2f} para estabilizar gradientes"
        )

        scaled_scores = attention_scores / np.sqrt(self.d_model)
        print_info("Escalonamento aplicado para evitar satura√ß√£o do softmax")

        # üìä PASSO 3: Aplicar Softmax
        print_step(3, "APLICANDO SOFTMAX PARA OBTER PESOS DE ATEN√á√ÉO", "üìä")

        print_math_explanation(
            "Attention_Weights = softmax(Scaled_Scores)",
            "Converte scores em probabilidades (soma = 1 para cada linha)"
        )

        attention_weights = softmax(scaled_scores, axis=-1)
        print_info("Softmax aplicado - cada linha agora soma 1.0")

        # Visualiza√ß√£o dos scores e weights
        fig, axes = plt.subplots(1, 2, figsize=(18, 8))
        fig.suptitle('üßÆ Transforma√ß√£o de Scores em Pesos de Aten√ß√£o', fontsize=16, fontweight='bold')

        # Attention Scores (antes do softmax)
        im1 = axes[0].imshow(scaled_scores, cmap='RdYlBu_r', aspect='auto', interpolation='nearest')
        axes[0].set_title('üßÆ Attention Scores (Escalados)\n"Compatibilidade bruta entre tokens"', fontweight='bold')
        axes[0].set_xlabel('Tokens (Keys)', fontweight='bold')
        axes[0].set_ylabel('Tokens (Queries)', fontweight='bold')
        axes[0].set_xticks(range(8))
        axes[0].set_yticks(range(8))
        axes[0].set_xticklabels(tokens, rotation=45, ha='right')
        axes[0].set_yticklabels(tokens)
        axes[0].grid(True, alpha=0.3)
        plt.colorbar(im1, ax=axes[0], label='Score Value')

        # Attention Weights (depois do softmax)
        im2 = axes[1].imshow(attention_weights, cmap='YlOrRd', aspect='auto', interpolation='nearest')
        axes[1].set_title('üìä Attention Weights (ap√≥s Softmax)\n"Probabilidades de aten√ß√£o"', fontweight='bold')
        axes[1].set_xlabel('Tokens (Keys)', fontweight='bold')
        axes[1].set_ylabel('Tokens (Queries)', fontweight='bold')
        axes[1].set_xticks(range(8))
        axes[1].set_yticks(range(8))
        axes[1].set_xticklabels(tokens, rotation=45, ha='right')
        axes[1].set_yticklabels(tokens)
        axes[1].grid(True, alpha=0.3)

        # Adicionar valores dos pesos na visualiza√ß√£o
        for i in range(8):
            for j in range(8):
                weight_val = attention_weights[i, j]
                color = "white" if weight_val > 0.15 else "black"
                axes[1].text(j, i, f'{weight_val:.2f}', ha="center", va="center",
                           color=color, fontsize=9, fontweight='bold')

        plt.colorbar(im2, ax=axes[1], label='Peso de Aten√ß√£o')
        plt.tight_layout()
        plt.show()

        # Verifica√ß√£o do softmax
        row_sums = attention_weights.sum(axis=1)
        print_info("VERIFICA√á√ÉO DO SOFTMAX:")
        print(f"  ‚Ä¢ Soma de cada linha: {row_sums}")
        print(f"  ‚Ä¢ Todas pr√≥ximas de 1.0? {np.allclose(row_sums, 1.0)}")

        # üéØ PASSO 4: Aplicar pesos aos Values
        print_step(4, "COMPUTANDO OUTPUT FINAL", "üéØ")

        print_math_explanation(
            "Output = Attention_Weights @ Values",
            "Combina√ß√£o ponderada dos values baseada nos pesos de aten√ß√£o"
        )

        output = attention_weights @ V
        print_info(f"Output computado: {output.shape}")
        print_info("Cada token agora cont√©m informa√ß√£o de todos os outros tokens relevantes")

        # Visualiza√ß√£o do processo final
        fig, axes = plt.subplots(1, 3, figsize=(22, 7))
        fig.suptitle('üéØ Processo Final: Attention √ó Values = Output', fontsize=16, fontweight='bold')

        # Values originais
        im1 = axes[0].imshow(V.T, cmap='Blues', aspect='auto', interpolation='nearest')
        axes[0].set_title('üíé Values Matrix\n"Informa√ß√£o dispon√≠vel"', fontweight='bold')
        axes[0].set_xlabel('Tokens', fontweight='bold')
        axes[0].set_ylabel('Dimens√µes', fontweight='bold')
        axes[0].set_xticks(range(8))
        axes[0].set_xticklabels(tokens, rotation=45, ha='right')
        axes[0].grid(True, alpha=0.3)
        plt.colorbar(im1, ax=axes[0], label='Valor')

        # Attention Weights
        im2 = axes[1].imshow(attention_weights, cmap='YlOrRd', aspect='auto', interpolation='nearest')
        axes[1].set_title('üìä Attention Weights\n"Quanto pegar de cada token"', fontweight='bold')
        axes[1].set_xlabel('Tokens (Keys)', fontweight='bold')
        axes[1].set_ylabel('Tokens (Queries)', fontweight='bold')
        axes[1].set_xticks(range(8))
        axes[1].set_yticks(range(8))
        axes[1].set_xticklabels(tokens, rotation=45, ha='right')
        axes[1].set_yticklabels(tokens)
        axes[1].grid(True, alpha=0.3)
        plt.colorbar(im2, ax=axes[1], label='Peso')

        # Output final
        im3 = axes[2].imshow(output.T, cmap='Purples', aspect='auto', interpolation='nearest')
        axes[2].set_title('üéØ Attention Output\n"Informa√ß√£o enriquecida"', fontweight='bold')
        axes[2].set_xlabel('Tokens', fontweight='bold')
        axes[2].set_ylabel('Dimens√µes', fontweight='bold')
        axes[2].set_xticks(range(8))
        axes[2].set_xticklabels(tokens, rotation=45, ha='right')
        axes[2].grid(True, alpha=0.3)
        plt.colorbar(im3, ax=axes[2], label='Output')

        plt.tight_layout()
        plt.show()

        # üìà An√°lise estat√≠stica
        print_info("AN√ÅLISE ESTAT√çSTICA DO OUTPUT:")
        print(f"  ‚Ä¢ Magnitude m√©dia do input: {np.mean(np.abs(X)):.4f}")
        print(f"  ‚Ä¢ Magnitude m√©dia do output: {np.mean(np.abs(output)):.4f}")
        print(f"  ‚Ä¢ Vari√¢ncia do input: {np.var(X):.4f}")
        print(f"  ‚Ä¢ Vari√¢ncia do output: {np.var(output):.4f}")

        return Q, K, V, attention_weights, output

    def analyze_attention_patterns(self, attention_weights):
        """
        üìà An√°lise detalhada e educativa dos padr√µes de aten√ß√£o
        """
        print_header("AN√ÅLISE DETALHADA DOS PADR√ïES DE ATEN√á√ÉO", "üìà")

        tokens = ['O', 'gato', 'subiu', 'no', 'telhado', 'ontem', '√†', 'noite']

        fig, axes = plt.subplots(2, 2, figsize=(18, 14))
        fig.suptitle('üìà An√°lise Aprofundada dos Padr√µes de Aten√ß√£o', fontsize=16, fontweight='bold')

        # 1Ô∏è‚É£ An√°lise espec√≠fica para "gato"
        print_step(1, "ANALISANDO ATEN√á√ÉO INDIVIDUAL", "üîé")

        gato_attention = attention_weights[1, :]  # Token "gato" (√≠ndice 1)
        bars1 = axes[0,0].bar(range(8), gato_attention, color='skyblue', alpha=0.8, edgecolor='navy')
        axes[0,0].set_title('üê± Aten√ß√£o de "gato" para outros tokens\n"Para onde o gato olha?"', fontweight='bold')
        axes[0,0].set_xlabel('Tokens', fontweight='bold')
        axes[0,0].set_ylabel('Peso de Aten√ß√£o', fontweight='bold')
        axes[0,0].set_xticks(range(8))
        axes[0,0].set_xticklabels(tokens, rotation=45, ha='right')
        axes[0,0].grid(True, alpha=0.3, axis='y')

        # Adicionar valores nas barras com melhor formata√ß√£o
        for i, (bar, v) in enumerate(zip(bars1, gato_attention)):
            height = bar.get_height()
            axes[0,0].text(bar.get_x() + bar.get_width()/2., height + 0.01,
                          f'{v:.3f}', ha='center', va='bottom', fontweight='bold')

        # 2Ô∏è‚É£ An√°lise espec√≠fica para "subiu" (verbo)
        subiu_attention = attention_weights[2, :]  # Token "subiu" (√≠ndice 2)
        bars2 = axes[0,1].bar(range(8), subiu_attention, color='lightcoral', alpha=0.8, edgecolor='darkred')
        axes[0,1].set_title('üèÉ Aten√ß√£o de "subiu" para outros tokens\n"O verbo conecta sujeito e objeto"', fontweight='bold')
        axes[0,1].set_xlabel('Tokens', fontweight='bold')
        axes[0,1].set_ylabel('Peso de Aten√ß√£o', fontweight='bold')
        axes[0,1].set_xticks(range(8))
        axes[0,1].set_xticklabels(tokens, rotation=45, ha='right')
        axes[0,1].grid(True, alpha=0.3, axis='y')

        for i, (bar, v) in enumerate(zip(bars2, subiu_attention)):
            height = bar.get_height()
            axes[0,1].text(bar.get_x() + bar.get_width()/2., height + 0.01,
                          f'{v:.3f}', ha='center', va='bottom', fontweight='bold')

        # 3Ô∏è‚É£ Mapa de calor completo com anota√ß√µes
        mask = np.zeros_like(attention_weights)
        mask[np.diag_indices_from(mask)] = True  # Mascarar diagonal principal

        sns.heatmap(attention_weights, annot=True, fmt='.3f',
                   xticklabels=tokens, yticklabels=tokens,
                   cmap='YlOrRd', ax=axes[1,0], cbar_kws={'label': 'Peso de Aten√ß√£o'},
                   linewidths=0.5, linecolor='white')
        axes[1,0].set_title('üî• Matriz de Aten√ß√£o Completa\n"Mapa de todas as rela√ß√µes"', fontweight='bold')
        axes[1,0].set_xlabel('Atendendo a (Keys)', fontweight='bold')
        axes[1,0].set_ylabel('Atendendo de (Queries)', fontweight='bold')

        # 4Ô∏è‚É£ Distribui√ß√£o estat√≠stica dos pesos
        weights_flat = attention_weights.flatten()
        n, bins, patches = axes[1,1].hist(weights_flat, bins=25, alpha=0.7, color='green',
                                         edgecolor='darkgreen', linewidth=1.2)
        axes[1,1].set_title('üìä Distribui√ß√£o dos Pesos de Aten√ß√£o\n"Como os pesos se distribuem?"', fontweight='bold')
        axes[1,1].set_xlabel('Valor do Peso de Aten√ß√£o', fontweight='bold')
        axes[1,1].set_ylabel('Frequ√™ncia', fontweight='bold')

        # Adicionar estat√≠sticas
        mean_weight = np.mean(weights_flat)
        median_weight = np.median(weights_flat)
        std_weight = np.std(weights_flat)

        axes[1,1].axvline(mean_weight, color='red', linestyle='--', linewidth=2,
                         label=f'M√©dia: {mean_weight:.3f}')
        axes[1,1].axvline(median_weight, color='blue', linestyle='--', linewidth=2,
                         label=f'Mediana: {median_weight:.3f}')
        axes[1,1].legend()
        axes[1,1].grid(True, alpha=0.3)

        plt.tight_layout()
        plt.show()

        # üîç An√°lise lingu√≠stica detalhada
        print_step(2, "AN√ÅLISE LINGU√çSTICA DOS PADR√ïES", "üß†")

        print_info("RELA√á√ïES SINT√ÅTICAS E SEM√ÇNTICAS DETECTADAS:")
        print("‚îÄ" * 60)

        # An√°lise das conex√µes mais fortes para cada token
        for i, token_query in enumerate(tokens):
            # Encontrar os 3 tokens com maior aten√ß√£o (excluindo self-attention)
            attention_row = attention_weights[i].copy()
            attention_row[i] = 0  # Remover self-attention para an√°lise
            top_indices = np.argsort(attention_row)[::-1][:3]

            print(f"\nüéØ '{token_query}' presta mais aten√ß√£o em:")
            for j, idx in enumerate(top_indices):
                weight = attention_weights[i, idx]
                relation_type = self._classify_relation(token_query, tokens[idx], weight)
                print(f"   {j+1}. '{tokens[idx]}' (peso: {weight:.3f}) - {relation_type}")

        # Estat√≠sticas globais
        print_info(f"\nESTAT√çSTICAS GLOBAIS:")
        print(f"  ‚Ä¢ Aten√ß√£o m√©dia: {mean_weight:.4f}")
        print(f"  ‚Ä¢ Desvio padr√£o: {std_weight:.4f}")
        print(f"  ‚Ä¢ Aten√ß√£o m√°xima: {np.max(attention_weights):.4f}")
        print(f"  ‚Ä¢ Aten√ß√£o m√≠nima: {np.min(attention_weights):.4f}")
        print(f"  ‚Ä¢ Esparsidade (pesos < 0.01): {np.sum(attention_weights < 0.01)/(8*8)*100:.1f}%")

        return attention_weights

    def _classify_relation(self, token1, token2, weight):
        """
        üè∑Ô∏è Classifica o tipo de rela√ß√£o entre dois tokens baseado no peso de aten√ß√£o
        """
        relations = {
            ("O", "gato"): "üë• Rela√ß√£o determinante-substantivo",
            ("gato", "subiu"): "üîó Rela√ß√£o sujeito-verbo",
            ("subiu", "telhado"): "üéØ Rela√ß√£o verbo-objeto",
            ("no", "telhado"): "üìç Rela√ß√£o preposi√ß√£o-objeto",
            ("ontem", "noite"): "‚è∞ Rela√ß√£o temporal",
            ("√†", "noite"): "üåô Express√£o temporal fixa"
        }

        key = (token1, token2)
        if key in relations:
            return relations[key]
        elif weight > 0.2:
            return "üî• Rela√ß√£o forte"
        elif weight > 0.1:
            return "‚ö° Rela√ß√£o moderada"
        else:
            return "üí´ Rela√ß√£o fraca"

# =============================================================================
# üéØ CLASSE MULTI-HEAD ATTENTION
# =============================================================================

class MultiHeadAttention:
    def __init__(self, d_model=64, num_heads=8):
        """
        üéØ Inicializa Multi-Head Attention

        M√∫ltiplas "cabe√ßas" de aten√ß√£o permitem ao modelo focar em diferentes
        tipos de rela√ß√µes simultaneamente (sint√°ticas, sem√¢nticas, posicionais).
        """
        self.d_model = d_model
        self.num_heads = num_heads
        self.d_k = d_model // num_heads  # Dimens√£o por cabe√ßa

        print_info(f"Multi-Head Attention inicializada:")
        print_info(f"  ‚Ä¢ N√∫mero de cabe√ßas: {num_heads}")
        print_info(f"  ‚Ä¢ Dimens√£o por cabe√ßa: {self.d_k}")
        print_info(f"  ‚Ä¢ Dimens√£o total: {d_model}")

        np.random.seed(42)
        # Pesos espec√≠ficos para cada cabe√ßa
        self.W_q_heads = [np.random.randn(d_model, self.d_k) * 0.1 for _ in range(num_heads)]
        self.W_k_heads = [np.random.randn(d_model, self.d_k) * 0.1 for _ in range(num_heads)]
        self.W_v_heads = [np.random.randn(d_model, self.d_k) * 0.1 for _ in range(num_heads)]

        # Matriz de proje√ß√£o final
        self.W_o = np.random.randn(d_model, d_model) * 0.1

        print_info("Matrizes de peso inicializadas para todas as cabe√ßas")

    def single_head_attention(self, X, W_q, W_k, W_v, head_num):
        """
        üéØ Computa aten√ß√£o para uma √∫nica cabe√ßa
        """
        Q = X @ W_q
        K = X @ W_k
        V = X @ W_v

        attention_scores = Q @ K.T / np.sqrt(self.d_k)
        attention_weights = softmax(attention_scores, axis=-1)
        output = attention_weights @ V

        return output, attention_weights

    def compute_multi_head_attention(self, X):
        """
        üé≠ Computa multi-head attention com visualiza√ß√µes educativas
        """
        print_header("MULTI-HEAD ATTENTION - M√öLTIPLAS PERSPECTIVAS", "üé≠")

        print_math_explanation(
            "MultiHead(Q,K,V) = Concat(head‚ÇÅ,...,head‚Çï)W_O",
            "Cada cabe√ßa foca em diferentes aspectos, depois s√£o combinadas"
        )

        head_outputs = []
        head_attentions = []

        print_step(1, "COMPUTANDO CADA CABE√áA INDIVIDUALMENTE", "üéØ")

        # Computar cada cabe√ßa
        for i in range(self.num_heads):
            print(f"  üî∏ Processando cabe√ßa {i+1}/{self.num_heads}...")
            output, attention = self.single_head_attention(
                X, self.W_q_heads[i], self.W_k_heads[i], self.W_v_heads[i], i+1
            )
            head_outputs.append(output)
            head_attentions.append(attention)

        print_info(f"Todas as {self.num_heads} cabe√ßas computadas com sucesso!")

        # Concatenar outputs das cabe√ßas
        print_step(2, "CONCATENANDO OUTPUTS DAS CABE√áAS", "üîó")
        concatenated = np.concatenate(head_outputs, axis=-1)
        print_info(f"Shape ap√≥s concatena√ß√£o: {concatenated.shape}")

        # Proje√ß√£o final
        print_step(3, "APLICANDO PROJE√á√ÉO FINAL", "üéØ")
        final_output = concatenated @ self.W_o
        print_info(f"Shape do output final: {final_output.shape}")

        print_math_explanation(
            "Output_final = Concat(heads) @ W_O",
            "Proje√ß√£o linear combina informa√ß√µes de todas as cabe√ßas"
        )

        # Visualizar padr√µes das diferentes cabe√ßas
        self.visualize_multi_head_patterns(head_attentions)

        return final_output, head_attentions

    def visualize_multi_head_patterns(self, head_attentions):
        """
        üé® Visualiza padr√µes de aten√ß√£o de diferentes cabe√ßas
        """
        print_step(4, "VISUALIZANDO PADR√ïES DE CADA CABE√áA", "üé®")

        tokens = ['O', 'gato', 'subiu', 'no', 'telhado', 'ontem', '√†', 'noite']

        fig, axes = plt.subplots(2, 4, figsize=(24, 12))
        fig.suptitle('üé≠ Especializa√ß√£o das Cabe√ßas de Aten√ß√£o', fontsize=16, fontweight='bold')
        axes = axes.flatten()

        # Cores diferentes para cada cabe√ßa
        cmaps = ['Reds', 'Blues', 'Greens', 'Purples', 'Oranges', 'YlOrRd', 'YlGnBu', 'RdPu']

        for i in range(self.num_heads):
            im = axes[i].imshow(head_attentions[i], cmap=cmaps[i], aspect='auto', interpolation='nearest')
            axes[i].set_title(f'üéØ Cabe√ßa {i+1}\n{self._get_head_specialty(i)}', fontweight='bold')
            axes[i].set_xticks(range(8))
            axes[i].set_yticks(range(8))
            axes[i].set_xticklabels(tokens, rotation=45, ha='right', fontsize=9)
            axes[i].set_yticklabels(tokens, fontsize=9)
            axes[i].grid(True, alpha=0.3)
            plt.colorbar(im, ax=axes[i], shrink=0.8)

        plt.tight_layout()
        plt.show()

        # An√°lise de especializa√ß√£o
        print_step(5, "AN√ÅLISE DE ESPECIALIZA√á√ÉO DAS CABE√áAS", "üß†")
        print_info("PADR√ïES DETECTADOS EM CADA CABE√áA:")
        print("‚îÄ" * 50)

        for i in range(self.num_heads):
            attention = head_attentions[i]
            specialty = self._analyze_head_pattern(attention, i)
            print(f"\nüéØ Cabe√ßa {i+1}: {specialty['type']}")
            print(f"   üìä {specialty['description']}")
            print(f"   üìà M√©tricas: {specialty['metrics']}")

    def _get_head_specialty(self, head_num):
        """
        üè∑Ô∏è Retorna a especializa√ß√£o simulada de cada cabe√ßa
        """
        specialties = [
            "Rela√ß√µes Sint√°ticas",
            "Depend√™ncias Locais",
            "Conex√µes Sem√¢nticas",
            "Aten√ß√£o Posicional",
            "Padr√µes Globais",
            "Estrutura Temporal",
            "Rela√ß√µes Funcionais",
            "Aten√ß√£o Distributiva"
        ]
        return specialties[head_num % len(specialties)]

    def _analyze_head_pattern(self, attention, head_num):
        """
        üìä Analisa o padr√£o de aten√ß√£o de uma cabe√ßa espec√≠fica
        """
        # Calcular m√©tricas para classificar o padr√£o
        local_attention = 0
        distant_attention = 0
        diagonal_attention = np.trace(attention)

        for row in range(len(attention)):
            for col in range(len(attention[row])):
                distance = abs(row - col)
                if distance <= 1:
                    local_attention += attention[row, col]
                elif distance >= 3:
                    distant_attention += attention[row, col]

        entropy = -np.sum(attention * np.log(attention + 1e-10))
        sparsity = np.sum(attention < 0.05) / (8 * 8)

        # Classificar padr√£o baseado nas m√©tricas
        if local_attention > distant_attention * 1.5:
            pattern_type = "üîç PADR√ÉO LOCAL"
            description = "Foca em tokens adjacentes e rela√ß√µes pr√≥ximas"
        elif distant_attention > local_attention:
            pattern_type = "üåê PADR√ÉO GLOBAL"
            description = "Conecta tokens distantes e rela√ß√µes de longo alcance"
        elif diagonal_attention > 0.3:
            pattern_type = "üéØ PADR√ÉO AUTO-FOCADO"
            description = "Alta aten√ß√£o para o pr√≥prio token"
        else:
            pattern_type = "‚öñÔ∏è PADR√ÉO BALANCEADO"
            description = "Distribui aten√ß√£o uniformemente"

        metrics = f"Local: {local_attention:.2f}, Distante: {distant_attention:.2f}, Entropia: {entropy:.2f}"

        return {
            'type': pattern_type,
            'description': description,
            'metrics': metrics
        }

# =============================================================================
# üèóÔ∏è TRANSFORMER BLOCK COMPLETO
# =============================================================================

class TransformerBlock:
    def __init__(self, d_model=64, num_heads=8, d_ff=256):
        """
        üèóÔ∏è Inicializa um bloco Transformer completo

        Um bloco Transformer consiste em:
        1. Multi-Head Attention + Residual Connection + Layer Norm
        2. Feed Forward Network + Residual Connection + Layer Norm
        """
        self.d_model = d_model
        self.multi_head_attention = MultiHeadAttention(d_model, num_heads)

        print_info("Transformer Block inicializado:")
        print_info(f"  ‚Ä¢ Dimens√£o do modelo: {d_model}")
        print_info(f"  ‚Ä¢ N√∫mero de cabe√ßas: {num_heads}")
        print_info(f"  ‚Ä¢ Dimens√£o Feed-Forward: {d_ff}")

        # Feed Forward Network
        np.random.seed(42)
        self.W1 = np.random.randn(d_model, d_ff) * 0.1
        self.W2 = np.random.randn(d_ff, d_model) * 0.1
        self.b1 = np.zeros(d_ff)
        self.b2 = np.zeros(d_model)

        print_info("Redes Feed-Forward inicializadas")

    def layer_norm(self, x, epsilon=1e-6):
        """
        üîß Layer Normalization - estabiliza o treinamento
        """
        mean = np.mean(x, axis=-1, keepdims=True)
        var = np.var(x, axis=-1, keepdims=True)
        return (x - mean) / np.sqrt(var + epsilon)

    def feed_forward(self, x):
        """
        üß† Feed Forward Network com ativa√ß√£o ReLU
        """
        # Primeira camada linear + ReLU
        hidden = np.maximum(0, x @ self.W1 + self.b1)
        # Segunda camada linear
        output = hidden @ self.W2 + self.b2
        return output

    def forward(self, x):
        """
        üöÄ Forward pass completo do transformer block
        """
        print_header("TRANSFORMER BLOCK COMPLETO EM A√á√ÉO", "üöÄ")

        print_info("ARQUITETURA DO TRANSFORMER BLOCK:")
        print("  1Ô∏è‚É£ Multi-Head Attention")
        print("  2Ô∏è‚É£ + Residual Connection")
        print("  3Ô∏è‚É£ + Layer Normalization")
        print("  4Ô∏è‚É£ Feed Forward Network")
        print("  5Ô∏è‚É£ + Residual Connection")
        print("  6Ô∏è‚É£ + Layer Normalization")

        # Etapa 1: Multi-Head Attention
        print_step(1, "APLICANDO MULTI-HEAD ATTENTION", "üé≠")
        attn_output, head_attentions = self.multi_head_attention.compute_multi_head_attention(x)

        # Etapa 2: Primeira Residual Connection
        print_step(2, "PRIMEIRA RESIDUAL CONNECTION", "üîó")
        print_math_explanation(
            "x‚ÇÅ = x + MultiHeadAttention(x)",
            "Soma o input original com o output da aten√ß√£o (evita vanishing gradients)"
        )

        x_after_attn = x + attn_output
        print_info(f"Shape ap√≥s residual connection: {x_after_attn.shape}")

        # Etapa 3: Primeira Layer Normalization
        print_step(3, "PRIMEIRA LAYER NORMALIZATION", "‚öñÔ∏è")
        print_math_explanation(
            "LayerNorm(x) = Œ≥ * (x - Œº) / œÉ + Œ≤",
            "Normaliza cada exemplo independentemente (estabiliza treinamento)"
        )

        x_norm1 = self.layer_norm(x_after_attn)
        print_info(f"Shape ap√≥s layer norm 1: {x_norm1.shape}")

        # Etapa 4: Feed Forward Network
        print_step(4, "FEED FORWARD NETWORK", "üß†")
        print_math_explanation(
            "FFN(x) = max(0, xW‚ÇÅ + b‚ÇÅ)W‚ÇÇ + b‚ÇÇ",
            "Rede neural de 2 camadas com ReLU (processamento n√£o-linear)"
        )

        ff_output = self.feed_forward(x_norm1)
        print_info(f"Shape ap√≥s feed forward: {ff_output.shape}")

        # Etapa 5: Segunda Residual Connection
        print_step(5, "SEGUNDA RESIDUAL CONNECTION", "üîó")
        x_after_ff = x_norm1 + ff_output
        print_info(f"Shape ap√≥s segunda residual: {x_after_ff.shape}")

        # Etapa 6: Segunda Layer Normalization
        print_step(6, "SEGUNDA LAYER NORMALIZATION", "‚öñÔ∏è")
        final_output = self.layer_norm(x_after_ff)
        print_info(f"Shape do output final: {final_output.shape}")

        # Visualizar todas as transforma√ß√µes
        self.visualize_transformer_stages(x, x_after_attn, x_norm1, ff_output, final_output)

        return final_output, head_attentions

    def visualize_transformer_stages(self, x_input, x_attn, x_norm1, x_ff, x_final):
        """
        üé® Visualiza as diferentes etapas do transformer block
        """
        print_step(7, "VISUALIZA√á√ÉO DAS TRANSFORMA√á√ïES", "üé®")

        tokens = ['O', 'gato', 'subiu', 'no', 'telhado', 'ontem', '√†', 'noite']

        fig, axes = plt.subplots(1, 5, figsize=(28, 6))
        fig.suptitle('üîÑ Evolu√ß√£o das Representa√ß√µes no Transformer Block', fontsize=16, fontweight='bold')

        stages = [
            (x_input, 'üì• Input\nOriginal', 'RdBu_r'),
            (x_attn, 'üé≠ Ap√≥s\nAttention', 'Greens'),
            (x_norm1, '‚öñÔ∏è Ap√≥s\nLayerNorm 1', 'Blues'),
            (x_ff, 'üß† Ap√≥s\nFeed Forward', 'Purples'),
            (x_final, 'üéØ Output\nFinal', 'Oranges')
        ]

        for i, (data, title, cmap) in enumerate(stages):
            im = axes[i].imshow(data.T, cmap=cmap, aspect='auto', interpolation='nearest')
            axes[i].set_title(title, fontweight='bold', fontsize=11)
            axes[i].set_xlabel('Tokens', fontweight='bold')
            if i == 0:
                axes[i].set_ylabel('Dimens√µes', fontweight='bold')
            axes[i].set_xticks(range(8))
            axes[i].set_xticklabels(tokens, rotation=45, ha='right', fontsize=9)
            axes[i].grid(True, alpha=0.3)
            plt.colorbar(im, ax=axes[i], shrink=0.8)

        plt.tight_layout()
        plt.show()

        # An√°lise das transforma√ß√µes
        print_info("AN√ÅLISE DAS TRANSFORMA√á√ïES:")
        transformations = [
            ("Input ‚Üí Attention", x_input, x_attn),
            ("Attention ‚Üí LayerNorm", x_attn, x_norm1),
            ("LayerNorm ‚Üí FeedForward", x_norm1, x_ff),
            ("FeedForward ‚Üí Final", x_ff, x_final)
        ]

        for name, before, after in transformations:
            change_magnitude = np.linalg.norm(after - before)
            print(f"  ‚Ä¢ {name}: Magnitude da mudan√ßa = {change_magnitude:.4f}")

# =============================================================================
# üöÄ SIMULA√á√ÉO COMPLETA PRINCIPAL
# =============================================================================

def run_complete_simulation():
    """
    üöÄ Executa a simula√ß√£o completa com todas as explica√ß√µes educativas
    """
    print_header("INICIANDO SIMULA√á√ÉO EDUCATIVA COMPLETA", "üöÄ", 80)

    print_info("Esta simula√ß√£o demonstrar√° todos os conceitos fundamentais dos Transformers:")
    print("  üéØ Embeddings e Positional Encoding")
    print("  üîç Self-Attention Mechanism passo a passo")
    print("  üé≠ Multi-Head Attention")
    print("  üèóÔ∏è Transformer Block completo")
    print("  üìä An√°lises e compara√ß√µes")

    # Inicializar simulador
    print_step(1, "INICIALIZA√á√ÉO DO SIMULADOR", "‚öôÔ∏è")
    sim = TransformerSimulator(d_model=64, seq_length=8)

    # Etapa 1: Embeddings e Positional Encoding
    print_header("ETAPA 1: FUNDAMENTOS - EMBEDDINGS E POSI√á√ïES", "üìä")
    embeddings = sim.visualize_embeddings_and_positional()

    # Etapa 2: Self-Attention detalhado
    print_header("ETAPA 2: MECANISMO DE SELF-ATTENTION", "üîç")
    Q, K, V, attention_weights, attention_output = sim.compute_attention_step_by_step(embeddings)

    # Etapa 3: An√°lise de padr√µes
    print_header("ETAPA 3: AN√ÅLISE DOS PADR√ïES DE ATEN√á√ÉO", "üìà")
    sim.analyze_attention_patterns(attention_weights)

    # Etapa 4: Multi-Head Attention
    print_header("ETAPA 4: MULTI-HEAD ATTENTION", "üé≠")
    mha = MultiHeadAttention(d_model=64, num_heads=8)
    mha_output, head_attentions = mha.compute_multi_head_attention(embeddings)

    # Etapa 5: Transformer Block Completo
    print_header("ETAPA 5: TRANSFORMER BLOCK COMPLETO", "üèóÔ∏è")
    transformer_block = TransformerBlock(d_model=64, num_heads=8, d_ff=256)
    final_output, _ = transformer_block.forward(embeddings)

    # Etapa 6: Compara√ß√£o Final
    print_header("ETAPA 6: COMPARA√á√ÉO E RESULTADOS FINAIS", "üèÜ")
    visualize_final_comparison(embeddings, attention_output, mha_output, final_output)

    return sim, mha, transformer_block

def visualize_final_comparison(embeddings, single_attn, multi_attn, transformer_out):
    """
    üèÜ Visualiza√ß√£o final comparando todos os outputs
    """
    print_step(1, "COMPARA√á√ÉO VISUAL DOS RESULTADOS", "üé®")

    fig, axes = plt.subplots(2, 2, figsize=(18, 12))
    fig.suptitle('üèÜ Evolu√ß√£o Completa: Do Input ao Transformer Final', fontsize=16, fontweight='bold')

    tokens = ['O', 'gato', 'subiu', 'no', 'telhado', 'ontem', '√†', 'noite']

    stages = [
        (embeddings, 'üì• Input Original\n(Embeddings + Posi√ß√£o)', 'RdBu_r'),
        (single_attn, 'üîç Single-Head Attention\n(Uma perspectiva)', 'Greens'),
        (multi_attn, 'üé≠ Multi-Head Attention\n(M√∫ltiplas perspectivas)', 'Blues'),
        (transformer_out, 'üèóÔ∏è Transformer Block Final\n(Processamento completo)', 'Oranges')
    ]

    axes_flat = axes.flatten()

    for i, (data, title, cmap) in enumerate(stages):
        im = axes_flat[i].imshow(data.T, cmap=cmap, aspect='auto', interpolation='nearest')
        axes_flat[i].set_title(title, fontweight='bold', fontsize=12)
        axes_flat[i].set_xlabel('Tokens', fontweight='bold')
        axes_flat[i].set_ylabel('Dimens√µes', fontweight='bold')
        axes_flat[i].set_xticks(range(8))
        axes_flat[i].set_xticklabels(tokens, rotation=45, ha='right')
        axes_flat[i].grid(True, alpha=0.3)
        plt.colorbar(im, ax=axes_flat[i])

    plt.tight_layout()
    plt.show()

    # M√©tricas comparativas
    print_step(2, "M√âTRICAS COMPARATIVAS FINAIS", "üìä")

    metrics = {
        'Input Original': embeddings,
        'Single-Head Attention': single_attn,
        'Multi-Head Attention': multi_attn,
        'Transformer Final': transformer_out
    }

    print_info("AN√ÅLISE QUANTITATIVA:")
    print("‚îÄ" * 50)

    for name, data in metrics.items():
        variance = np.var(data)
        mean_abs = np.mean(np.abs(data))
        l2_norm = np.linalg.norm(data)

        print(f"\nüìà {name}:")
        print(f"   ‚Ä¢ Vari√¢ncia: {variance:.6f}")
        print(f"   ‚Ä¢ Magnitude m√©dia: {mean_abs:.6f}")
        print(f"   ‚Ä¢ Norma L2: {l2_norm:.6f}")

# =============================================================================
# üéì RESUMO EDUCACIONAL E INSIGHTS
# =============================================================================

def educational_summary():
    """
    üéì Resumo educacional abrangente dos conceitos
    """
    print_header("RESUMO EDUCACIONAL COMPLETO", "üéì", 80)

    print_info("CONCEITOS FUNDAMENTAIS DEMONSTRADOS:")

    concepts = {
        "1Ô∏è‚É£ EMBEDDINGS + POSITIONAL ENCODING": {
            "conceito": "Representa√ß√£o num√©rica de tokens com informa√ß√£o posicional",
            "detalhes": [
                "‚Ä¢ Tokens ‚Üí vetores densos (embeddings sem√¢nticos)",
                "‚Ä¢ Posi√ß√£o codificada via fun√ß√µes seno/cosseno",
                "‚Ä¢ Permite paraleliza√ß√£o (n√£o precisa de recorr√™ncia)",
                "‚Ä¢ Base fundamental para todo o processamento"
            ],
            "formula": "Final_Embedding = Token_Embedding + Positional_Encoding"
        },

        "2Ô∏è‚É£ SELF-ATTENTION MECHANISM": {
            "conceito": "Cada token 'conversa' com todos os outros para entender contexto",
            "detalhes": [
                "‚Ä¢ Query: 'O que este token est√° procurando?'",
                "‚Ä¢ Key: 'O que este token pode oferecer?'",
                "‚Ä¢ Value: 'Que informa√ß√£o este token carrega?'",
                "‚Ä¢ Attention Score mede compatibilidade Q¬∑K",
                "‚Ä¢ Softmax normaliza em probabilidades",
                "‚Ä¢ Output = soma ponderada dos Values"
            ],
            "formula": "Attention(Q,K,V) = softmax(QK^T/‚àöd_k)V"
        },

        "3Ô∏è‚É£ MULTI-HEAD ATTENTION": {
            "conceito": "M√∫ltiplas 'perspectivas' de aten√ß√£o processadas em paralelo",
            "detalhes": [
                "‚Ä¢ Cada cabe√ßa foca em aspectos diferentes",
                "‚Ä¢ Captura rela√ß√µes sint√°ticas E sem√¢nticas",
                "‚Ä¢ Processamento paralelo ‚Üí efici√™ncia",
                "‚Ä¢ Outputs concatenados e projetados",
                "‚Ä¢ Permite especializa√ß√£o autom√°tica"
            ],
            "formula": "MultiHead(Q,K,V) = Concat(head‚ÇÅ,...,head‚Çï)W_O"
        },

        "4Ô∏è‚É£ TRANSFORMER BLOCK": {
            "conceito": "Bloco completo com aten√ß√£o, feed-forward e normaliza√ß√µes",
            "detalhes": [
                "‚Ä¢ Multi-Head Attention processa rela√ß√µes",
                "‚Ä¢ Residual Connections previnem vanishing gradients",
                "‚Ä¢ Layer Normalization estabiliza treinamento",
                "‚Ä¢ Feed-Forward adiciona n√£o-linearidade",
                "‚Ä¢ Arquitetura empilh√°vel para modelos profundos"
            ],
            "formula": "Block(x) = LayerNorm(x + FFN(LayerNorm(x + MHA(x))))"
        }
    }

    for title, info in concepts.items():
        print(f"\n{title}")
        print("‚ïê" * len(title))
        print(f"üéØ CONCEITO: {info['conceito']}")
        print(f"üìê F√ìRMULA: {info['formula']}")
        print("üìù DETALHES:")
        for detail in info['detalhes']:
            print(f"   {detail}")

    print_header("PRINCIPAIS INSIGHTS E CONCLUS√ïES", "üí°")

    insights = [
        "üöÄ PARALELIZA√á√ÉO: Attention permite processar toda sequ√™ncia simultaneamente",
        "üé≠ ESPECIALIZA√á√ÉO: Multi-head captura diferentes tipos de rela√ß√µes automaticamente",
        "üîó RESIDUAL CONNECTIONS: Permitem treinar redes muito profundas sem degrada√ß√£o",
        "‚öñÔ∏è LAYER NORM: Estabiliza treinamento e acelera converg√™ncia",
        "üß† ESCALABILIDADE: Arquitetura funciona de pequenos modelos at√© LLMs gigantes",
        "üéØ INTERPRETABILIDADE: Pesos de aten√ß√£o mostram 'racioc√≠nio' do modelo",
        "üí´ VERSATILIDADE: Mesma arquitetura para tradu√ß√£o, gera√ß√£o, classifica√ß√£o...",
        "üèóÔ∏è FOUNDATION: Base dos modelos mais avan√ßados (GPT, BERT, T5, etc.)"
    ]

    for insight in insights:
        print(f"   {insight}")

    print_header("COMO USAR ESTE CONHECIMENTO", "üõ†Ô∏è")

    print_info("APLICA√á√ïES PR√ÅTICAS:")
    applications = [
        "üìö EDUCA√á√ÉO: Entender como LLMs 'pensam' e processam linguagem",
        "üîß DEBUGGING: Interpretar pesos de aten√ß√£o para debugar modelos",
        "üé® CRIATIVIDADE: Projetar arquiteturas especializadas para tarefas espec√≠ficas",
        "üìä AN√ÅLISE: Usar attention maps para explicar decis√µes do modelo",
        "‚ö° OTIMIZA√á√ÉO: Identificar gargalos e otimizar performance",
        "üîç PESQUISA: Base para desenvolver novas arquiteturas e melhorias"
    ]

    for app in applications:
        print(f"   {app}")

# =============================================================================
# üéÆ DEMONSTRA√á√ïES INTERATIVAS AVAN√áADAS
# =============================================================================

def interactive_attention_demo():
    """
    üéÆ Demonstra√ß√£o interativa avan√ßada com exemplos pr√°ticos
    """
    print_header("DEMONSTRA√á√ïES INTERATIVAS AVAN√áADAS", "üéÆ")

    # Exemplo 1: Resolu√ß√£o de Ambiguidade
    print_step(1, "RESOLU√á√ÉO DE AMBIGUIDADE CONTEXTUAL", "üß©")

    print_info("EXEMPLO: Como o modelo diferencia 'banco' em contextos diferentes")

    # Simular dois contextos diferentes
    contexts = {
        'financeiro': {
            'tokens': ['Fui', 'ao', 'banco', 'sacar', 'dinheiro', 'ontem', 'pela', 'manh√£'],
            'attention_pattern': create_financial_context_attention()
        },
        'rio': {
            'tokens': ['Sentei', 'no', 'banco', 'do', 'rio', 'para', 'pescar', 'tranquilo'],
            'attention_pattern': create_river_context_attention()
        }
    }

    visualize_context_comparison(contexts)

    # Exemplo 2: An√°lise de Depend√™ncias Sint√°ticas
    print_step(2, "CAPTURA DE DEPEND√äNCIAS SINT√ÅTICAS", "üå≥")
    demonstrate_syntactic_dependencies()

    # Exemplo 3: Aten√ß√£o Temporal
    print_step(3, "PROCESSAMENTO DE INFORMA√á√ïES TEMPORAIS", "‚è∞")
    demonstrate_temporal_attention()

def create_financial_context_attention():
    """
    üè¶ Cria padr√£o de aten√ß√£o para contexto financeiro
    """
    # Simula√ß√£o de padr√£o onde 'banco' se conecta fortemente com 'sacar', 'dinheiro'
    pattern = np.array([
        [0.15, 0.1, 0.3, 0.2, 0.15, 0.05, 0.03, 0.02],  # Fui
        [0.1, 0.2, 0.4, 0.1, 0.1, 0.05, 0.03, 0.02],   # ao
        [0.05, 0.1, 0.25, 0.35, 0.2, 0.03, 0.01, 0.01], # banco -> sacar, dinheiro
        [0.05, 0.05, 0.4, 0.2, 0.25, 0.03, 0.01, 0.01], # sacar -> banco, dinheiro
        [0.02, 0.05, 0.3, 0.25, 0.3, 0.05, 0.02, 0.01], # dinheiro -> banco, sacar
        [0.1, 0.05, 0.15, 0.1, 0.1, 0.3, 0.1, 0.1],     # ontem
        [0.05, 0.05, 0.1, 0.05, 0.05, 0.2, 0.25, 0.25], # pela
        [0.02, 0.03, 0.05, 0.03, 0.05, 0.15, 0.25, 0.42] # manh√£
    ])

    # Normalizar para que cada linha some 1
    return pattern / pattern.sum(axis=1, keepdims=True)

def create_river_context_attention():
    """
    üèûÔ∏è Cria padr√£o de aten√ß√£o para contexto do rio
    """
    # Simula√ß√£o onde 'banco' se conecta com 'rio', 'pescar'
    pattern = np.array([
        [0.2, 0.1, 0.25, 0.15, 0.1, 0.1, 0.05, 0.05],   # Sentei
        [0.1, 0.15, 0.35, 0.2, 0.1, 0.05, 0.03, 0.02],  # no
        [0.05, 0.15, 0.2, 0.25, 0.25, 0.05, 0.03, 0.02], # banco -> rio, pescar
        [0.02, 0.1, 0.3, 0.15, 0.35, 0.05, 0.02, 0.01],  # do
        [0.02, 0.05, 0.35, 0.15, 0.25, 0.1, 0.05, 0.03], # rio -> banco, pescar
        [0.05, 0.05, 0.2, 0.1, 0.15, 0.2, 0.15, 0.1],    # para
        [0.03, 0.05, 0.25, 0.1, 0.2, 0.15, 0.2, 0.02],   # pescar -> banco, rio
        [0.1, 0.1, 0.15, 0.1, 0.1, 0.1, 0.15, 0.2]       # tranquilo
    ])

    return pattern / pattern.sum(axis=1, keepdims=True)

def visualize_context_comparison(contexts):
    """
    üé® Visualiza a compara√ß√£o entre diferentes contextos
    """
    fig, axes = plt.subplots(1, 2, figsize=(20, 8))
    fig.suptitle('üß© Resolu√ß√£o de Ambiguidade: "banco" em Contextos Diferentes',
                 fontsize=16, fontweight='bold')

    for i, (context_name, context_data) in enumerate(contexts.items()):
        tokens = context_data['tokens']
        pattern = context_data['attention_pattern']

        # Visualizar matriz de aten√ß√£o
        im = axes[i].imshow(pattern, cmap='YlOrRd', aspect='auto', interpolation='nearest')

        title = f'üè¶ Contexto Financeiro' if context_name == 'financeiro' else f'üèûÔ∏è Contexto Natural'
        axes[i].set_title(f'{title}\n"{" ".join(tokens)}"', fontweight='bold')

        axes[i].set_xticks(range(len(tokens)))
        axes[i].set_yticks(range(len(tokens)))
        axes[i].set_xticklabels(tokens, rotation=45, ha='right')
        axes[i].set_yticklabels(tokens)
        axes[i].grid(True, alpha=0.3)

        # Destacar conex√µes do 'banco'
        banco_idx = tokens.index('banco')
        banco_attention = pattern[banco_idx]

        # Encontrar as conex√µes mais fortes
        top_connections = np.argsort(banco_attention)[-3:]
        for j in top_connections:
            if j != banco_idx:  # N√£o destacar self-attention
                axes[i].add_patch(Rectangle((j-0.4, banco_idx-0.4), 0.8, 0.8,
                                          fill=False, edgecolor='red', linewidth=3))

        plt.colorbar(im, ax=axes[i])

    plt.tight_layout()
    plt.show()

    # An√°lise quantitativa
    print_info("AN√ÅLISE QUANTITATIVA DA DIFERENCIA√á√ÉO:")
    banco_financial = contexts['financeiro']['attention_pattern'][2]  # 'banco' no contexto financeiro
    banco_river = contexts['rio']['attention_pattern'][2]  # 'banco' no contexto do rio

    print("  üè¶ Contexto Financeiro - 'banco' presta aten√ß√£o em:")
    financial_top = np.argsort(banco_financial)[-3:][::-1]
    for idx in financial_top:
        if idx != 2:  # N√£o incluir self-attention
            token = contexts['financeiro']['tokens'][idx]
            weight = banco_financial[idx]
            print(f"     ‚Ä¢ '{token}' (peso: {weight:.3f})")

    print("\n  üèûÔ∏è Contexto Natural - 'banco' presta aten√ß√£o em:")
    river_top = np.argsort(banco_river)[-3:][::-1]
    for idx in river_top:
        if idx != 2:  # N√£o incluir self-attention
            token = contexts['rio']['tokens'][idx]
            weight = banco_river[idx]
            print(f"     ‚Ä¢ '{token}' (peso: {weight:.3f})")

def demonstrate_syntactic_dependencies():
    """
    üå≥ Demonstra como o modelo captura depend√™ncias sint√°ticas
    """
    print_info("EXEMPLO: Captura de rela√ß√µes sint√°ticas na frase:")
    print_info("'A menina que estava correndo trope√ßou na pedra'")

    tokens = ['A', 'menina', 'que', 'estava', 'correndo', 'trope√ßou', 'na', 'pedra']

    # Simular padr√µes sint√°ticos
    syntactic_pattern = np.array([
        [0.2, 0.7, 0.05, 0.02, 0.01, 0.01, 0.005, 0.005],  # A -> menina (det-subst)
        [0.1, 0.3, 0.1, 0.05, 0.05, 0.35, 0.03, 0.02],     # menina -> trope√ßou (subj-verb)
        [0.02, 0.3, 0.2, 0.25, 0.2, 0.02, 0.005, 0.005],   # que -> estava, correndo
        [0.01, 0.15, 0.15, 0.25, 0.4, 0.02, 0.01, 0.01],   # estava -> correndo (aux-verb)
        [0.01, 0.2, 0.1, 0.3, 0.35, 0.02, 0.01, 0.01],     # correndo -> estava, menina
        [0.02, 0.35, 0.05, 0.05, 0.05, 0.3, 0.1, 0.08],    # trope√ßou -> menina, pedra
        [0.005, 0.02, 0.01, 0.01, 0.01, 0.1, 0.2, 0.655],  # na -> pedra (prep-obj)
        [0.005, 0.1, 0.01, 0.02, 0.02, 0.25, 0.15, 0.455]  # pedra -> trope√ßou
    ])

    fig, axes = plt.subplots(1, 2, figsize=(18, 8))
    fig.suptitle('üå≥ Captura de Depend√™ncias Sint√°ticas', fontsize=16, fontweight='bold')

    # Matriz de aten√ß√£o
    im1 = axes[0].imshow(syntactic_pattern, cmap='Greens', aspect='auto', interpolation='nearest')
    axes[0].set_title('üå≥ Matriz de Depend√™ncias Sint√°ticas', fontweight='bold')
    axes[0].set_xticks(range(len(tokens)))
    axes[0].set_yticks(range(len(tokens)))
    axes[0].set_xticklabels(tokens, rotation=45, ha='right')
    axes[0].set_yticklabels(tokens)
    axes[0].grid(True, alpha=0.3)
    plt.colorbar(im1, ax=axes[0])

    # Grafo de depend√™ncias
    import networkx as nx

    G = nx.DiGraph()
    for i, token in enumerate(tokens):
        G.add_node(i, label=token)

    # Adicionar arestas para rela√ß√µes fortes (> 0.25)
    for i in range(len(tokens)):
        for j in range(len(tokens)):
            if syntactic_pattern[i, j] > 0.25 and i != j:
                G.add_edge(i, j, weight=syntactic_pattern[i, j])

    pos = nx.spring_layout(G, k=3, iterations=50)

    axes[1].set_title('üîó Grafo de Depend√™ncias\n(conex√µes > 0.25)', fontweight='bold')

    # Desenhar componentes do grafo
    nx.draw_networkx_nodes(G, pos, ax=axes[1], node_color='lightgreen',
                          node_size=1500, alpha=0.8)

    edges = G.edges()
    weights = [G[u][v]['weight'] for u, v in edges]
    nx.draw_networkx_edges(G, pos, ax=axes[1], width=[w*8 for w in weights],
                          alpha=0.7, edge_color='darkgreen', arrows=True, arrowsize=25)

    labels = {i: tokens[i] for i in range(len(tokens))}
    nx.draw_networkx_labels(G, pos, labels, ax=axes[1], font_size=10, font_weight='bold')

    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

    # An√°lise das depend√™ncias
    print_info("DEPEND√äNCIAS SINT√ÅTICAS IDENTIFICADAS:")
    dependencies = [
        ("A ‚Üí menina", "Determinante-Substantivo", syntactic_pattern[0,1]),
        ("menina ‚Üí trope√ßou", "Sujeito-Verbo", syntactic_pattern[1,5]),
        ("estava ‚Üí correndo", "Auxiliar-Verbo Principal", syntactic_pattern[3,4]),
        ("na ‚Üí pedra", "Preposi√ß√£o-Objeto", syntactic_pattern[6,7]),
        ("trope√ßou ‚Üí pedra", "Verbo-Objeto Indireto", syntactic_pattern[5,7])
    ]

    for relation, type_rel, weight in dependencies:
        print(f"  üîó {relation}: {type_rel} (peso: {weight:.3f})")

def demonstrate_temporal_attention():
    """
    ‚è∞ Demonstra como o modelo processa informa√ß√µes temporais
    """
    print_info("EXEMPLO: Processamento de sequ√™ncia temporal:")
    print_info("'Ontem choveu, hoje faz sol, amanh√£ pode nevar'")

    tokens = ['Ontem', 'choveu', 'hoje', 'faz', 'sol', 'amanh√£', 'pode', 'nevar']

    # Simular aten√ß√£o temporal - marcadores temporais se conectam
    temporal_pattern = np.array([
        [0.25, 0.3, 0.2, 0.1, 0.05, 0.08, 0.01, 0.01],   # Ontem -> choveu, hoje
        [0.4, 0.25, 0.15, 0.05, 0.05, 0.05, 0.03, 0.02], # choveu -> ontem
        [0.15, 0.1, 0.25, 0.25, 0.15, 0.08, 0.01, 0.01], # hoje -> faz, sol
        [0.05, 0.03, 0.35, 0.25, 0.3, 0.01, 0.005, 0.005], # faz -> hoje, sol
        [0.02, 0.05, 0.25, 0.35, 0.25, 0.05, 0.02, 0.01], # sol -> hoje, faz
        [0.08, 0.05, 0.15, 0.05, 0.05, 0.25, 0.25, 0.12], # amanh√£ -> pode, nevar
        [0.02, 0.02, 0.05, 0.02, 0.02, 0.35, 0.25, 0.27], # pode -> amanh√£, nevar
        [0.01, 0.02, 0.03, 0.01, 0.02, 0.3, 0.3, 0.31]    # nevar -> amanh√£, pode
    ])

    fig, axes = plt.subplots(1, 2, figsize=(18, 8))
    fig.suptitle('‚è∞ Processamento de Informa√ß√µes Temporais', fontsize=16, fontweight='bold')

    # Matriz temporal
    im1 = axes[0].imshow(temporal_pattern, cmap='Blues', aspect='auto', interpolation='nearest')
    axes[0].set_title('‚è∞ Aten√ß√£o Temporal\n"Conectando eventos no tempo"', fontweight='bold')
    axes[0].set_xticks(range(len(tokens)))
    axes[0].set_yticks(range(len(tokens)))
    axes[0].set_xticklabels(tokens, rotation=45, ha='right')
    axes[0].set_yticklabels(tokens)
    axes[0].grid(True, alpha=0.3)

    # Destacar marcadores temporais
    temporal_markers = [0, 2, 5]  # ontem, hoje, amanh√£
    for marker in temporal_markers:
        axes[0].add_patch(Rectangle((marker-0.4, marker-0.4), 0.8, 0.8,
                                  fill=False, edgecolor='red', linewidth=3))

    plt.colorbar(im1, ax=axes[0])

    # An√°lise por per√≠odo temporal
    periods = ['Passado\n(Ontem choveu)', 'Presente\n(Hoje faz sol)', 'Futuro\n(Amanh√£ pode nevar)']
    period_indices = [(0, 1), (2, 3, 4), (5, 6, 7)]

    attention_by_period = []
    for indices in period_indices:
        period_attention = 0
        count = 0
        for i in indices:
            for j in indices:
                if i != j:
                    period_attention += temporal_pattern[i, j]
                    count += 1
        attention_by_period.append(period_attention / count if count > 0 else 0)

    bars = axes[1].bar(periods, attention_by_period, color=['red', 'green', 'blue'], alpha=0.7)
    axes[1].set_title('üìä Coes√£o Interna por Per√≠odo\n"Quanto cada per√≠odo se conecta internamente"', fontweight='bold')
    axes[1].set_ylabel('Aten√ß√£o M√©dia Interna', fontweight='bold')
    axes[1].grid(True, alpha=0.3, axis='y')

    # Adicionar valores nas barras
    for bar, value in zip(bars, attention_by_period):
        height = bar.get_height()
        axes[1].text(bar.get_x() + bar.get_width()/2., height + 0.01,
                    f'{value:.3f}', ha='center', va='bottom', fontweight='bold')

    plt.tight_layout()
    plt.show()

    # An√°lise temporal
    print_info("AN√ÅLISE TEMPORAL:")
    print("  üìÖ Marcadores temporais identificados:")
    temporal_analysis = [
        ("Ontem", "Passado", 0, temporal_pattern[0,1]),
        ("Hoje", "Presente", 2, temporal_pattern[2,3] + temporal_pattern[2,4]),
        ("Amanh√£", "Futuro", 5, temporal_pattern[5,6] + temporal_pattern[5,7])
    ]

    for marker, time_ref, idx, connection_strength in temporal_analysis:
        print(f"     ‚Ä¢ {marker} ({time_ref}): Conex√£o com eventos = {connection_strength:.3f}")

# =============================================================================
# üìö FUN√á√ÉO PRINCIPAL DE EXECU√á√ÉO
# =============================================================================

def main():
    """
    üìö Fun√ß√£o principal que executa toda a simula√ß√£o educativa
    """
    print_header("SIMULA√á√ÉO EDUCATIVA COMPLETA DE TRANSFORMERS", "üéì", 90)

    print_info("üéØ OBJETIVOS DESTA SIMULA√á√ÉO:")
    objectives = [
        "Compreender cada componente dos Transformers",
        "Visualizar como a aten√ß√£o funciona na pr√°tica",
        "Analisar padr√µes lingu√≠sticos capturados",
        "Comparar diferentes arquiteturas de aten√ß√£o",
        "Desenvolver intui√ß√£o sobre o funcionamento interno"
    ]

    for i, obj in enumerate(objectives, 1):
        print(f"   {i}. {obj}")

    print("\n" + "üöÄ INICIANDO SIMULA√á√ÉO..." + "\n")

    # Executar simula√ß√£o principal
    simulator, multi_head, transformer = run_complete_simulation()

    # Demonstra√ß√µes interativas
    interactive_attention_demo()

    # Resumo educacional
    educational_summary()

    print_header("SIMULA√á√ÉO CONCLU√çDA COM SUCESSO!", "üéâ", 90)

    print_info("PR√ìXIMOS PASSOS SUGERIDOS:")
    next_steps = [
        "üî¨ Experimente modificar os par√¢metros (d_model, num_heads)",
        "üìù Teste com suas pr√≥prias frases de exemplo",
        "üé® Crie visualiza√ß√µes personalizadas para seus casos",
        "üèóÔ∏è Implemente varia√ß√µes da arquitetura Transformer",
        "üìä Analise attention patterns em modelos reais",
        "üß† Estude papers fundamentais sobre Transformers"
    ]

    for step in next_steps:
        print(f"   {step}")

    print_info("RECURSOS ADICIONAIS:")
    resources = [
        "üìñ Paper original: 'Attention Is All You Need' (Vaswani et al.)",
        "üé• Videos educativos sobre Deep Learning",
        "üíª Implementa√ß√µes em PyTorch/TensorFlow",
        "üîç Ferramtas de visualiza√ß√£o: Bertviz, Attention-viz",
        "üìö Cursos online: CS224N, Fast.ai, Hugging Face"
    ]

    for resource in resources:
        print(f"   {resource}")

    return simulator, multi_head, transformer

# =============================================================================
# üéØ INSTRU√á√ïES PARA EXECU√á√ÉO
# =============================================================================

"""
üöÄ COMO EXECUTAR ESTA SIMULA√á√ÉO:

1Ô∏è‚É£ GOOGLE COLAB (Recomendado):
   ‚Ä¢ Abra colab.research.google.com
   ‚Ä¢ Crie um novo notebook
   ‚Ä¢ Cole este c√≥digo completo
   ‚Ä¢ Execute: main()

2Ô∏è‚É£ JUPYTER NOTEBOOK:
   ‚Ä¢ Instale: pip install numpy matplotlib seaborn scipy pandas networkx
   ‚Ä¢ Abra Jupyter Notebook
   ‚Ä¢ Cole o c√≥digo e execute

3Ô∏è‚É£ PYTHON SCRIPT:
   ‚Ä¢ Salve como transformer_simulation.py
   ‚Ä¢ Execute: python transformer_simulation.py

üìä SA√çDAS ESPERADAS:
‚úÖ 15+ visualiza√ß√µes educativas
‚úÖ Explica√ß√µes passo a passo
‚úÖ An√°lises quantitativas
‚úÖ Demonstra√ß√µes interativas
‚úÖ Resumo educacional completo

‚è±Ô∏è TEMPO DE EXECU√á√ÉO: ~3-5 minutos
üíæ MEM√ìRIA NECESS√ÅRIA: ~2GB RAM
üéØ N√çVEL: Intermedi√°rio a Avan√ßado

üéì P√öBLICO-ALVO:
‚Ä¢ Estudantes de IA/ML
‚Ä¢ Pesquisadores em NLP
‚Ä¢ Desenvolvedores de LLMs
‚Ä¢ Curiosos sobre Transformers

üî• CARACTER√çSTICAS √öNICAS:
‚Ä¢ Explica√ß√µes visuais detalhadas
‚Ä¢ C√≥digo educativo e comentado
‚Ä¢ Exemplos pr√°ticos de uso
‚Ä¢ An√°lises interpret√°veis
‚Ä¢ Foco pedag√≥gico
"""

# Execute a simula√ß√£o completa
if __name__ == "__main__":
    main()
```
